```python
import pandas as pd
# pandas is aliased as pd
import numpy as np
# numpy is aliased as np
import matplotlib.pyplot as pt
# pyplot is aliased as plt
from sklearn.preprocessing import StandardScaler
```

```python
df.isnull().sum()
```

```python
# Load the dataset
df = pd.read_csv('CAR DETAILS (2).csv')
```

```python
# 1. One-Hot Encoding categorical columns
df_encoded = pd.get_dummies(df, columns=['fuel', 'seller_type', 'transmission', 'owner'], drop_first=True)
```

```python
# 2. Scaling numerical columns: 'year', 'selling_price', 'km_driven'
scaler = StandardScaler()
```

```python
df_encoded[['year', 'selling_price', 'km_driven']] = scaler.fit_transform(df_encoded[['year', 'selling_price', 'km_driven']])
```

```python
# Display the processed data
print(df_encoded.head())
```

```python
# Save the processed data to a new CSV file
df_encoded.to_csv('CAR_DETAILS_PROCESSED.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'CAR DETAILS (2).csv'  # Provide the correct path to your CSV file
car_data = pd.read_csv('CAR DETAILS (2).CSV')
```

```python
# Plot 1: Distribution of car manufacturing year
plt.figure(figsize=(10, 6))
sns.histplot(car_data['year'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Car Manufacturing Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Plot 2: Distribution of car prices
plt.figure(figsize=(10, 6))
sns.histplot(car_data['selling_price'], bins=30, kde=True, color='green')
plt.title('Distribution of Selling Price', fontsize=16)
plt.xlabel('Selling Price (INR)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Plot 3: Fuel type count
plt.figure(figsize=(8, 5))
sns.countplot(y='fuel', data=car_data, palette='Set2', order=car_data['fuel'].value_counts().index)
plt.title('Count of Cars by Fuel Type', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Fuel Type', fontsize=12)
plt.show()
import warnings
warnings.filterwarnings("ignore")
```
