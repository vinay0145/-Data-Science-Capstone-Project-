```python
What is the average age of customers?
```

```python
import pandas as pd

data = pd.read_csv("Mall_Customers.csv")

avg_age = data["Age"].mean()
print("Average Age:", avg_age)
```

```python
What is the average annual income of customers?
```

```python
avg_income = data["Annual Income (k$)"].mean()
print("Average Income:", avg_income)
```

```python
Which gender spends more?
```

```python
spending_gender = data.groupby("Genre")["Spending Score (1-100)"].mean()
print(spending_gender)
```

```python
Which age group spends more?
```

```python
data['Age Group'] = pd.cut(data['Age'], bins=[0,20,40,60,100],
                          labels=['0-20','20-40','40-60','60+'])

age_spending = data.groupby("Age Group")["Spending Score (1-100)"].mean()
print(age_spending)
```

```python
Who are the top 5 customers by spending score?
```

```python
top_customers = data.sort_values(by="Spending Score (1-100)", ascending=False).head(5)
print(top_customers)
```
