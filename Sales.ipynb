```python
1.What is the Total Sales Revenue?
```

```python
import pandas as pd

data = pd.read_csv("sales_data_sample.csv", encoding='latin1')

total_sales = data["SALES"].sum()
print("Total Sales Revenue:", total_sales)
```

```python
2.Which Product Line Generates the Highest Sales?
```

```python
sales_product = data.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False)
print(sales_product)
```

```python
3.Which Country Has the Highest Sales
```

```python
sales_country = data.groupby("COUNTRY")["SALES"].sum().sort_values(ascending=False)
print(sales_country)
```

```python
4.Who Are the Top 10 Customers by Sales?
```

```python
top_customers = data.groupby("CUSTOMERNAME")["SALES"].sum().sort_values(ascending=False).head(10)
print(top_customers)
```

```python
5.What is the Monthly Sales Trend?
```

```python
monthly_sales = data.groupby("MONTH_ID")["SALES"].sum().sort_values()
print(monthly_sales)
```

```python
import matplotlib.pyplot as plt

monthly_sales.plot(kind="line")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
